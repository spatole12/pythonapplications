```python
#Making a basic bokeh line graph

#importing bokeh . . . figure acts like a container object
from bokeh.plotting import figure
from bokeh.io import output_file,show

#prepare some data. . . x and y ought to have same length
x=[1,2,3,4,5]
y=[6,7,8,9,10]

#prepare the output file
output_file("Line.html")

# create object of figure class
f=figure()

#Create line plot
f.line(x,y)
f.circle(x,y)
# f.triangle(x,y)

#write the plot in the figure object
show(f)
```

Bokeh with pandas

```python
#Making a basic bokeh line graph

#importing bokeh and pandas. . . figure acts like a container object
from bokeh.plotting import figure
from bokeh.io import output_file,show
import pandas

#prepare some data. . . x and y ought to have same length
df = pandas.read_csv("data.csv")
x=df["x"]
y=df["y"]

#prepare the output file
output_file("Line_from_csv.html")

# create object of figure class
f=figure()

#Create line plot
f.line(x,y)
f.circle(x,y)
# f.triangle(x,y)

#write the plot in the figure object
show(f)
```

Plotting weather data : Lec 194

```python
import pandas
from bokeh.plotting import figure
from bokeh.io import output_file,show

df2 = pandas.read_excel("http://pythonhow.com/data/verlegenhuken.xlsx",sheet_name=0)
df2["Temperature"] = df2["Temperature"]/10
df2["Pressure"] = df2["Pressure"]/10

p=figure(plot_width=500,plot_height=400, tools='pan')
 
p.title.text="Temperature and Air Pressure"
p.title.text_color="Gray"
p.title.text_font="times"
p.title.text_font_style="bold"
p.xaxis.minor_tick_line_color=None
p.yaxis.minor_tick_line_color=None
p.xaxis.axis_label="Temperature(Â°C)"
p.yaxis.axis_label="Pressure(hPa)"    
 

output_file("PracticeLec194.html")

p.circle(df2["Temperature"],df2["Pressure"],size=0.5)

show(p)
```

More visual attributes

```python
from bokeh.plotting import figure, output_file, show
p = figure(plot_width=500, plot_height=400, tools = 'pan, reset')
p.title.text = "Earthquakes"
p.title.text_color = "Orange"
p.title.text_font = "times"
p.title.text_font_style = "italic"
p.yaxis.minor_tick_line_color = "Yellow"
p.xaxis.axis_label = "Times"
p.yaxis.axis_label = "Value"
p.circle([1,2,3,4,5], [5,6,5,5,3], size = [i*2 for i in [8,12,14,15,20]], color="red", alpha=0.5)
output_file("Scatter_plotting.html")
show(p)
```
