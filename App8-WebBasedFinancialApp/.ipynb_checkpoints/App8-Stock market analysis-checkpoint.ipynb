```python
from pandas_datareader import data
import datetime
from bokeh.plotting import figure,output_file,show
from bokeh.embed import components
```

```python
data.DataReader?
```

```python
# stock symbols/tickers... Google it..AAPL
start = datetime.datetime(2015,11,1)
end = datetime.datetime(2016,3,10)

df = data.DataReader(name="AAPL",data_source="yahoo",start=start,end=end)
print(type(data.DataReader(name="AAPL",data_source="yahoo",start=start,end=end)))
```

```python
df
```

```python
date_increase = df.index[df.Close > df.Open]
date_decrease = df.index[df.Close < df.Open]
print(type(date_increase))
```

```python
p = figure(x_axis_type="datetime",width=1000,height=300)
p.title.text = "CandleStick Chart"
p.grid.grid_line_alpha = 0.3

hours_12 = 12*60*60*1000
p.rect(date_increase,(df.Open+df.Close)/2,hours_12,abs(df.Open-df.Close),fill_color="green",line_color="blue")
p.rect(date_decrease,(df.Open+df.Close)/2,hours_12,abs(df.Open-df.Close),fill_color="red",line_color="orange")

p.segment(df.index,df.High,df.index,df.Low,color="black")

output_file("Stock_analysis.html")

show(p)
```

METHOD:2

```python
def inc_decr(close,open):
    if close>open:
        value = "Increase"
    elif close<open:
        value="Decrease"
    else:
        value="Equal"
    return value

    
df["Status"] = [inc_decr(c,o) for c,o in zip(df.Close,df.Open)]
df["Middle"] = (df.Close+df.Open)/2
df["Height"] = abs(df.Close-df.Open)
```

```python
df
```

```python
p = figure(x_axis_type="datetime",width=1000,height=300,sizing_mode="scale_width")
p.title.text = "CandleStick Chart"

hours_12 = 12*60*60*1000
p.segment(df.index,df.High,df.index,df.Low,color="black")
p.rect(df.index[df.Status == "Increase"],df.Middle[df.Status=="Increase"],hours_12,df.Height[df.Status=="Increase"],fill_color="#CCFFFF",line_color="blue")
p.rect(df.index[df.Status == "Decrease"],df.Middle[df.Status=="Decrease"],hours_12,df.Height[df.Status=="Decrease"],fill_color="red",line_color="orange")
# output_file("Stock_analysis.html")

# show(p)

script1,div1 = components(p)
```
